```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
    
#The sigmoid function transforms the continuous real number into a range of ( 0 , 1 ), 
# so that the input value of the next layer is within a fixed range and the weight is more stable.
# Sigmoid functions are often used in the output layer of a binary classification.

def relu(x):
    return np.maximum(0,x)
    
# The Rectified Linear Unit is the most commonly used activation function in deep learning models.
# for any positive value x it returns that value back. And for negative values of x it returns 0.
# ReLU is chiefly implemented in hidden layers of Neural network. 


def tanh(x):
    return np.tanh(x)
    # return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))    #another formula that might be used.

# Tanh Activation is an activation function used for neural networks: f ( x )=(e^x)−(e^−x)/ (e^x) +(e^−x). 
# Historically, the tanh function became preferred over the sigmoid function as it gave better performance for multi-layer neural networks.
# Tanh functions are usually used in hidden layers of a neural network.

# Note: Activation functions like sigmoid, tanh, and ReLU are used in deep learning(ANN) to give neural networks non-linear properties

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x))

# The softmax activation function transforms the raw outputs of the neural network into a vector of probabilities, 
# essentially a probability distribution over the input classes
# softmax is used in multi-class classification.
```

```python
x=np.linspace(-10,10,100)
```

```python
fix,axs=plt.subplots(2,2,figsize=(8,8))
axs[0,0].plot(x,sigmoid(x))
axs[0,0].set_title("Sigmoid")
axs[0,1].plot(x,relu(x))
axs[0,1].set_title("ReLu")
axs[1,0].plot(x,tanh(x))
axs[1,0].set_title("tanh")
axs[1,1].plot(x,softmax(x))
axs[1,1].set_title("Softmax")
fix.suptitle("Common Activation Functions")
# sup represents superscript
plt.show()
```
